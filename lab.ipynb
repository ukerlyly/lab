```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import zeros

# Initial variables
L = 0.1
N = 100
dx = L/N
dt = 0.001
cCu = 385.0
KCu = 385.0
rhoCu = 8900.0
t = 0.0
tf = 50.0
alpha = KCu/(rhoCu*cCu)
counter = 0
colorindex = -1
steady_state_reached = False  # add the steady symbol
steady_state_time = None     # record steady time

# define the whole time
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 
          'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

# initial temerature array
Tx = zeros(N+1, float)
Tx_prev = zeros(N+1, float)  
Uxt = []

# set temperature
Tx[0] = 50
Tx[1:N] = 25
Tx[N] = 0
Tx_prev[:] = Tx[:]  


Uxt.append(Tx.copy())


ymin = Tx.min() - 1
ymax = Tx.max() + 1
x = np.linspace(0, L, N+1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].set_aspect(0.01)
axes[0].set_xlabel('Position (m)')
axes[0].tick_params(bottom=True, labelbottom=True, left=False, labelleft=False)
pcm = axes[0].pcolormesh(np.linspace(0, L, N+2), [0, 1], [Tx], 
                        cmap=plt.get_cmap('inferno'), vmin=ymin, vmax=ymax)
cbar = fig.colorbar(pcm, ax=axes[0])
cbar.ax.set_ylabel('Temperature (C)')


axes[1].plot(x, Tx, label=f"t = {t:.1f}s.")
axes[1].set_title('Scenario 2 Timesteps')
axes[1].set_xlabel('Position (m)')
axes[1].set_ylabel('Temperature (C)')
axes[1].set_ylim([ymin, ymax])
axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)


while t < tf:

    Tx_prev[:] = Tx[:]
    
    
    T_new = Tx.copy()
    for i in range(1, N):
        
        d2T_dx2 = (Tx[i+1] - 2*Tx[i] + Tx[i-1]) / dx**2
        T_new[i] = Tx[i] + alpha * d2T_dx2 * dt
    
    Tx = T_new
    
    
    Tx[0] = 50  
    Tx[N] = 0   
    
    
    Uxt.append(Tx.copy())
    
    
    t += dt
    counter += 1
    
    
    if counter % 100 == 0 and not steady_state_reached:
        # calculate the change of temp
        max_temp_change = np.max(np.abs(Tx - Tx_prev))
        
        # if the temp<0.01
        if max_temp_change < 0.001:  #0.001C
            steady_state_reached = True
            steady_state_time = t
            print(f"System reached steady state at t = {steady_state_time:.2f}s")
    
    # update
    if counter % 100 == 0:
        
        pcm.set_array([Tx])
        
       
        ln, = axes[1].plot(x, Tx, color=colors[(colorindex + 1) % len(colors)], 
                          alpha=0.7, label=f"t = {t:.1f}s.")
        
       
        if steady_state_reached and steady_state_time >= t - dt and steady_state_time <= t:
            ln.set_color('red')
            ln.set_linewidth(2.5)
            ln.set_label(f"Steady State (t = {steady_state_time:.1f}s)")
           
            axes[1].scatter(x[50], Tx[50], color='red', s=50, zorder=5)
        
        
        axes[0].set_title(f"Distribution at t = {t:.1f}s.")
        
      
        plt.pause(0.01)
        
        
        if counter > 500:  
            ln.remove()
        
        colorindex += 1

#final temperature
print(f"Final temperature at midpoint: {Tx[50]:.2f}C")
if steady_state_time is None:
    print("System did not reach steady state within simulation time.")
else:
    print(f"Steady state reached at t = {steady_state_time:.2f}s")


Uxt_array = np.array(Uxt)
time_points = np.linspace(0, tf, len(Uxt_array))
position_points = np.linspace(0, L, N+1)
 
fig2, ax2 = plt.subplots(figsize=(10, 6))
pcol = ax2.pcolormesh(position_points, time_points, Uxt_array, 
                     cmap=plt.get_cmap('inferno'), shading='auto')
cbar2 = fig2.colorbar(pcol)
cbar2.ax.set_ylabel('Temperature (C)')
ax2.set_title('Temperature Distribution Over Time Scenario2')
ax2.set_xlabel('Position (m)')
ax2.set_ylabel('Time (s)')


if steady_state_time is not None:
    ax2.axhline(y=steady_state_time, color='lime', linestyle='--', linewidth=2)
    ax2.text(0.05, steady_state_time + 0.5, f'Steady State (t={steady_state_time:.1f}s)', 
             color='lime', fontsize=10, backgroundcolor='black')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Material properties (Copper)
L = 0.1          # Length of rod (m)
N = 100          # Number of spatial points
dx = L/N
dt = 0.001       # Time step (s)
KCu = 385.0      # Thermal conductivity (W/(mK))
rhoCu = 8900.0   # Density (kg/m)
cCu = 385.0      # Specific heat (J/(kgK))
alpha = KCu/(rhoCu*cCu)  # Thermal diffusivity (m/s)
tf = 50.0        # Final simulation time (s)

# Create position array
x = np.linspace(0, L, N+1)

def simulate_scenario(scenario):
    """Simulate heat diffusion for a given scenario"""
    # Initialize temperature array
    Tx = np.zeros(N+1, float)
    
    # Set initial conditions based on scenario
    if scenario == 1:
        Tx = 20 * np.sin(np.pi * x / L)
        bc_left = 0
        bc_right = 0
    elif scenario == 2:
        Tx[0] = 50
        Tx[1:N] = 25
        Tx[N] = 0
        bc_left = 50
        bc_right = 0
    elif scenario == 3:
        Tx[0:50] = 20
        Tx[50:70] = 100
        Tx[70:N+1] = 50
        bc_left = 20
        bc_right = 50
    elif scenario == 4:
        Tx[0:20] = 50
        Tx[20:40] = 10
        Tx[40:60] = 40
        Tx[60:80] = 20
        Tx[80:N+1] = 30
        bc_left = 50
        bc_right = 30
    
    # Store initial state
    T_initial = Tx.copy()
    
    # Prepare for time evolution
    t = 0.0
    T_short = None
    T_steady = None
    
    while t < tf:
        # Calculate second derivative
        d2T = np.zeros_like(Tx)
        d2T[1:-1] = (Tx[2:] - 2*Tx[1:-1] + Tx[:-2]) / dx**2
        
        # Update temperature (FTCS scheme)
        Tx = Tx + alpha * d2T * dt
        
        # Apply boundary conditions
        Tx[0] = bc_left
        Tx[-1] = bc_right
        
        # Store specific time states
        if t >= 0.2 and T_short is None:
            T_short = Tx.copy()
        
        t += dt
    
    T_steady = Tx.copy()
    
    return T_initial, T_short, T_steady

# Create figure with subplots
plt.figure(figsize=(14, 10))

# Simulate and plot all scenarios
for i, scenario in enumerate(range(1, 5)):
    T_init, T_short, T_steady = simulate_scenario(scenario)
    
    plt.subplot(2, 2, i+1)
    plt.plot(x, T_init, 'k:', label='Initial (t=0)')
    plt.plot(x, T_short, 'b-', linewidth=2, label='t0.2s')
    plt.plot(x, T_steady, 'r-', linewidth=2, label='Steady State')
    
    plt.title(f'Scenario {scenario}')
    plt.xlabel('Position (m)')
    plt.ylabel('Temperature (C)')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.ylim(-5, 110)

plt.tight_layout()
plt.show()
```
